# use DataFrame.apply()

```python
import pandas as pd
data={
    'data':['10/2/2011','11/2/2011','12/2/2011','13/2/2011'],
    'event':['Music','Poetry','Theatre','comedy'],
    'cost':[10000,5000,15000,20000]
    }
df=pd.DataFrame(data)
df
```

```python
df['Discounted_price']=df.apply(lambda a:a.cost-(a.cost*0.1),axis=1)
df
```
