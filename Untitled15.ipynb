```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("d:\cars.csv")
df
```

```python
plt.scatter(df['wt'],df['mpg'],color='green',marker='o')
plt.xlabel('wt')
plt.ylabel('mpg')
plt.title('scatter plot')
plt.show()
```

```python
range=(0,10)
bins=10
plt.hist(df['disp'],bins,range,color="red",histtype='bar',rwidth=0.8)
plt.xlabel('disp')
plt.ylabel('frequency')
plt.title('my histogram')
plt.show()
```

```python
plt.boxplot(df['hp'])
plt.ylabel('hp')
plt.title('my boxplot')
plt.show()
```
