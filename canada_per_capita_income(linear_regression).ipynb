```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:\\Users\\Hp\\OneDrive\\Documents\\Dataset\\canada_per_capita_income.csv')
df
```

```python
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df['year'],df['per capita income (US$)'],color='green',marker='*')
plt.show()
```

```python
new_df=df.drop('per capita income (US$)',axis='columns')
new_df
```

```python
capita=df['per capita income (US$)']
capita
```

```python
#creating a linear regression object
reg=linear_model.LinearRegression()
reg.fit(new_df,capita)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
